```python
def cat_onehot_cols(multcolumns):
    df_train_final=df_train
    i=0
    for fields in multcolumns:
        print(fields)
        df=pd.get_dummies(df_train[fields],drop_first=True)
        df_train_final.drop([fields],axis=1, inplace=True)
        if i==0:
            df_train_final=df.copy()
        else:
            df_final=pd.concat([df_train_final,df],axis=1)
        i=i+1
    df_train_final=pd.concat([df_train, df_train_final], axis=1)
    return df_train_final
    
```
