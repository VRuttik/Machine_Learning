```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df = pd.read_csv('/workspaces/codespaces-jupyter/data/train.csv')

# Deletion
# To drop rows with any missing values
df_cleaned1 = df.dropna()
print(df_cleaned1)

# To drop columns with any missing values
df_cleaned2 = df.dropna(axis=1)
print(df_cleaned2)
```

```python
# Imputation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df = pd.read_csv('/workspaces/codespaces-jupyter/data/train.csv')

# Impute missing values with mean
df_cleaned1 = df.fillna(df.mean())
print(df_cleaned1)

# Impute missing values with forward-fill
df_cleaned2 = df.fillna(method='ffill')
print(df_cleaned2)

# Imput missing values with backward-fill
df_cleaned3 = df.fillna(method='bfill')
print(df_cleaned3)
```
